# The electromagnetic spectrum (EM spectrum)

In your physics classes, you may have already learnt about the electromagnetic radiation. Electromagnetic energy, produced by the vibration of charged particles, travels in the form of waves through the atmosphere and the vacuum of space.The full range of EM radiation is called the electromagnetic spectrum. This spectrum is split into different bands with characteristic wavelengths and frequencies, e.g. (from short to long wavelength) gamma rays, X-Rays, Ultraviolet, visible light, infrared and radio waves. 
A large wavelength means a lower frequency, e.g. radio waves have very long wavelengths, but very low frequency. A wavelength is measured as one complete cycle of the wave, e.g. the distance from wave crest to wave crest.
Visible light sits in the middle of that range of long to shortwave radiation. Within the ranges of visible light, violet/blue light has the shortest wavelength (around 400 nanometers), whereas red light has the longest (700 nanometers). This small portion of energy is all that the human eye is able to detect.

# How is electromagnetic radiation important to satellites?

As mentioned above, we can only see a small portion of the electromagnetic spectrum, which we have called the visible light. We can build instruments which can act like our eyes and measure light in the visible range, and even detect all other forms of electromagnetic energy that we wouldn't otherwise be able to see. 

We have various instruments on satellites orbiting around the Earth, with the ability to measure different parts of the EM spectrum. They are integral parts of our communication systems, navigation systems and remote sensing of the Earth System. 

Some waves are absorbed or reflected by atmospheric components, like water vapor and carbon dioxide, while some wavelengths allow for unimpeded movement through the atmosphere; visible light has wavelengths that can be transmitted through the atmosphere. Microwave energy has wavelengths that can pass through clouds, an attribute utilized by instruments aboard many weather and communication platforms. The figures below show the opacity of the atmosphere (how much radiation is absorbed or transmitted). There are several "windows" of wavelengths where the atmosphere is trnasparent enough to let through the radiation at that wavelength.

The first figure shows the contrast between the spectrum of radiation measures at the top of the atmosphere - primarily the light from our Sun. However, by the time this light reaches the surface, the energy at certain wavelengths has been attenuated, so there is much less light arriving at the surface than we would expect if the Earth had no atmosphere. 

The primary source of the energy observed by instruments aboard space-based platforms is the Sun. The amount of the Sun's energy that is reflected depends on the roughness of the Earth's surface and its albedo, which is how well a surface reflects light instead of absorbing it. Snow, for example, has a very high albedo and reflects up to 90% of incoming solar radiation regardless of its colour/wavelength (this is why it looks bright white in satellite images). The ocean, on the other hand, reflects only about 6% of incoming solar radiation and absorbs the rest (this is why it looks so dark on satellite images). Often, when energy is absorbed, it is re-emitted, usually at longer wavelengths. For example, the energy absorbed by the ocean gets re-emitted as infrared radiation.

The second figure shows the absorption lines of some important gases in the atmosphere, as well as the total combined absorption of the atmosphere. When designing an instrument and choosing the spectral abnds and spectral resolutions, the scientists need to be fully aware of these absorption lines and atmospheric windows as they will heavily influence what the instrument can be used to study. For example, if the scientist wants to study the surface of the Earth, clouds will make it impossible to see the surface at certain wavelengths (eg. most of infrared radiation).
